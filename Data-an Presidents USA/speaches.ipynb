```python
import pandas as pd
from IPython.display import display_html
import string

trump = []
biden = []
obama = []
with open('trump.txt',encoding="utf8") as file:  
    for line in file:        
        for word in line.split():
            trump.append(word)
            
with open('Biden.txt',encoding="utf8") as file:  
    for line in file:        
        for word in line.split():
            biden.append(word)
            
with open('Obama.txt',encoding="utf8") as file:  
    for line in file:        
        for word in line.split():
            obama.append(word)

def organize(list):
    
    df = pd.DataFrame({'Words': list})
    df['Words'] = df['Words'].str.lower()#lower case
    df['Words'] = df['Words'].apply(lambda x:''.join([i for i in x if i not in string.punctuation]))#punctuation
    df['Lenght'] = df['Words'].apply(lambda x: len(x))#lenght words
    df.drop(df[df.Lenght < 5].index, inplace=True)#delete <5 words
    df['Frequency'] = df['Words'].map(df['Words'].value_counts())#count frequency
    df = df.sort_values(by=['Frequency'], ascending = False)#descending order
    df = df.drop_duplicates()#deleting duplicates
    df = df.head(20)
    #pd.set_option('display.max_rows', df.shape[0]+1)#to show every row in output
    return df

df1 = organize(trump)
df1.columns = ['Trump','Lenght','Frequency']
df2 = organize(biden)
df2.columns = ['Biden','Lenght','Frequency']
df3 = organize(obama)
df3.columns = ['Obama','Lenght','Frequency']


df1_styler = df1.style.set_table_attributes("style='display:inline'").set_caption('df1')
df2_styler = df2.style.set_table_attributes("style='display:inline'").set_caption('df2')
df3_styler = df3.style.set_table_attributes("style='display:inline'").set_caption('df3')
display_html(df1_styler._repr_html_()+df2_styler._repr_html_()+df3_styler._repr_html_(), raw=True)
```
