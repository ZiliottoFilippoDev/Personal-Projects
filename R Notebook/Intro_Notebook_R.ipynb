In R we don't need to specify which type of plot we want in general it will automaticcaly recognize the type. For example categorical or qauntitative variables

```r
#plot(dataset $ (use this variable) Species)
plot(iris$Species)
```

```r
plot(iris$Petal.Length)
```

```r
plot(iris$Species, iris$Petal.Width)
```

To plot the entire dataset

```r
plot(iris)
```

```r
plot(iris$Petal.Length, iris$Petal.Width,
    col = '#cc0000', #color red
    pch = 19, #Use solid circles for points
    main = 'Iris: Petal Length vs. Petal Width',
    xlab = 'Petal Length',
    ylab = 'Petal Width'   
    )
```

To plot math formulas

```r
plot(dnorm, -2*pi, 2*pi)
```

```r
plot(dnorm, -3, +3,
     col = 'red',
     lwd = 5,
     main = 'Standard Normal Distribution',
     xlab = 'Z-scores',
     ylab = 'Density'
     )
```
