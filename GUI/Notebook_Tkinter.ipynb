### Build a calculator

```python
root = Tk()
root.title('Simple Calculator')

e = Entry(root, width=35, borderwidth=5)

e.grid(row=0, column=0,
       columnspan=3, #this column has 3 buttons
       padx=10, pady=10)

def button_click(number):
    current = e.get()
    e.delete(0, END)
    e.insert(0, str(current) + str(number))
    
def button_clear():
    e.delete(0, END)
    
def button_add():
    global num, math
    math = 'addition'
    tmp1 = e.get()
    num = tmp1
    e.delete(0, END)
    
def button_equal():
    tmp2 = e.get()
    e.delete(0, END)
    
    if math == 'addition':
        e.insert(0, int(num) + int(tmp2))
    if math == 'subtraction':
        e.insert(0, int(num) - int(tmp2))
    if math == 'multiplication':
        e.insert(0, int(num) * int(tmp2))
    if math == 'division':
        e.insert(0, int(num) / int(tmp2))
    
def button_subtract():
    global num, math
    math = 'subtraction'
    tmp1 = e.get()
    num = tmp1
    e.delete(0, END)
    
def button_multiply():
    global num, math
    math = 'multiplication'
    tmp1 = e.get()
    num = tmp1
    e.delete(0, END)

def button_divide():
    global num, math
    math = 'division'
    tmp1 = e.get()
    num = tmp1
    e.delete(0, END)
    
#Define buttons
button_1 = Button(root, text='1', padx=40, pady='20', command=lambda :button_click(1))
button_2 = Button(root, text='2', padx=40, pady='20', command=lambda :button_click(2))
button_3 = Button(root, text='3', padx=40, pady='20', command=lambda :button_click(3))
button_4 = Button(root, text='4', padx=40, pady='20', command=lambda :button_click(4))
button_5 = Button(root, text='5', padx=40, pady='20', command=lambda :button_click(5))
button_6 = Button(root, text='6', padx=40, pady='20', command=lambda :button_click(6))
button_7 = Button(root, text='7', padx=40, pady='20', command=lambda :button_click(7))
button_8 = Button(root, text='8', padx=40, pady='20', command=lambda :button_click(8))
button_9 = Button(root, text='9', padx=40, pady='20', command=lambda :button_click(9))
button_0 = Button(root, text='0', padx=40, pady='20', command=lambda :button_click(0))
button_add = Button(root, text='+', padx=39, pady=20, command=button_add)
button_equal = Button(root, text='=', padx=87, pady=20, command=button_equal)
button_clear = Button(root, text='Clear', padx=79, pady=20, command=button_clear)

button_subtract = Button(root, text='-', padx=41, pady=20, command=button_subtract)
button_multiply = Button(root, text='*', padx=41, pady=20, command=button_multiply)
button_divide = Button(root, text='/', padx=41, pady=20, command=button_divide)

#Display buttons
button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)
button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)
button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)
button_0.grid(row=4, column=0)

button_clear.grid(row=4, column=1, columnspan=2)
button_add.grid(row=5, column=0)
button_equal.grid(row=5, column=1, columnspan=2)

button_subtract.grid(row=6, column=0)
button_multiply.grid(row=6, column=1)
button_divide.grid(row=6, column=2)




root.mainloop()
```

## Images, Icons, Exit buttons

```python
from PIL import ImageTk, Image

root = Tk()
root.title('My Tkinter program')
root.iconbitmap(r'Images/cat_icon.ico') #icon on the top left


my_img1 = ImageTk.PhotoImage(Image.open(r'Images/py_image.png'))
my_img3 = ImageTk.PhotoImage(Image.open(r'Images/poke2.jpg'))
my_img4 = ImageTk.PhotoImage(Image.open(r'Images/poke3.jpg'))

image_list = [my_img1, my_img3, my_img4]

my_label = Label(image=my_img1)
my_label.grid(row=0, column=0, columnspan=3)

#defining functions

def forward(image_number):
    global my_label, button_back, button_forward
    
    my_label.grid_forget()
    my_label = Label(image=image_list[image_number-1])
    button_forward = Button(root, text='>>',
                            command=lambda: forward(image_number+1))
    button_back = Button(root, text='<<', 
                         command=lambda: back(image_number-1))
    
    if image_number == len(image_list):
        button_forward = Button(root, text='>>', state=DISABLED)
    
    my_label.grid(row=0, column=0, columnspan=3)
    button_back.grid(row=1, column=0)
    button_forward.grid(row=1, column=2)
    
    #adding status bar
    text =('Image '+str(image_number)+ ' of ' + str(len(image_list)))
    status = Label(root, text=text,
               bd=1, 
               relief = SUNKEN,
               anchor=E
              ) 
    status.grid(row=2, column=0, columnspan=3, sticky=W+E)

def back(image_number):
    global my_label, button_back, button_forward
    
    my_label.grid_forget()
    my_label = Label(image=image_list[image_number-1])
    button_forward = Button(root, text='>>',
                            command=lambda: forward(image_number+1))
    button_back = Button(root, text='<<', 
                         command=lambda: back(image_number-1))
    if image_number == 1:
        button_back = Button(root, text='<<', state=DISABLED)
        
    my_label.grid(row=0, column=0, columnspan=3)
    button_back.grid(row=1, column=0)
    button_forward.grid(row=1, column=2)
    
    #adding status bar
    text =('Image '+str(image_number)+ ' of ' + str(len(image_list)))
    status = Label(root, text=text,
               bd=1, 
               relief = SUNKEN,
               anchor=E
              ) 
    status.grid(row=2, column=0, columnspan=3, sticky=W+E)




button_back = Button(root, text='<<', command=lambda: back(0))
button_exit = Button(root, text='Exit', command=root.destroy)
button_forward = Button(root, text='>>', command=lambda: forward(2))

my_label.grid(row=0, column=0, columnspan=3)
button_back.grid(row=1, column=0)
button_forward.grid(row=1, column=2, pady=10)
button_exit.grid(row=1, column=1)

text =('Image 1 of ' + str(len(image_list)))
status = Label(root, text=text,
               bd=1, 
               relief = SUNKEN,
               anchor=E
              ) 
status.grid(row=2, column=0, columnspan=3, sticky=W+E)


root.mainloop()
```

## Message Boxes

There are many types of boxes: showinfo, showwarning, showerror, askquestion, askokcancel, askyesno

```python
from tkinter import messagebox

root = Tk()

root.title('My Tkinter program')
root.iconbitmap(r'Images/cat_icon.ico')

def popup():
    response = messagebox.askokcancel('This is a popup','Hello World')
    Label(root, text=response).pack()
    
    if response == 0:
        Label(root, text='ANNULLA').pack()
    else:
        Label(root, text='OK').pack()
    

Button(root, text='Popup', command=popup).pack()

root.mainloop()
```

## Open file dialog
Use the function filename that returns the location of the seleted object

```python
from tkinter import filedialog
from PIL import ImageTk, Image

root = Tk()
root.title('My Tkinter program')
root.iconbitmap(r'Images/cat_icon.ico')

def open():
    global my_image
    root.filename = filedialog.askopenfilename(
    initialdir=r'Images',#indicate directory path
    title='Select a File',
    filetypes=(('png','*.png'),#open any name .png
               ('All files','*.*'),('PDF','.pdf'))#open any file any .
    )
    my_label = Label(root, text=root.filename).pack()
    my_image = ImageTk.PhotoImage(Image.open(root.filename))
    my_label = Label(root, image=my_image).pack()
    
btn = Button(root, text='Open File',command=open).pack()

root.mainloop()
```

## Graphing and Charts 
Actually since we are in jupyter notebook the graph is shown on the output of the cell. Run this program in python terminal to open the graph in another tkinter window

```python
from mpl_toolkits.mplot3d import axes3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

root = Tk()
root.title('My Tkinter program')
root.iconbitmap(r'Images/cat_icon.ico')
root.geometry('400x600')

#graphing a histogram
def graph():
    
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    X, Y, Z = axes3d.get_test_data(0.05)

    cset = ax.contour(X, Y, Z, extend3d=True, cmap=cm.coolwarm)
    ax.clabel(cset, fontsize=9, inline=True)
    plt.show()

my_button = Button(root, text='View Graph', command=graph)
my_button.pack()

root.mainloop()
```
